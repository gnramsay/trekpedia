```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

MAIN_URL="https://en.wikipedia.org/wiki/Star_Trek"

result = requests.get(MAIN_URL)

result.status_code

# result.text

bs = BeautifulSoup(result.text, 'lxml')
```

```python
pretty = bs.prettify()
print(pretty)
```

```python
with open("dump.html", "w") as f:
    f.write(pretty)
```

```python
# try a different way, more generic just using the '#Television' anchor...
# this works and is better I think.
tv = bs.find('a', href='#Television')
names = tv.parent.find_all('i')
# years = tv.parent.find_all('span', attrs={"class":"toctext"})
series = []
for name in names:
    series.append(name.text)
```

```python
series
```

```python
df = pd.DataFrame(series, columns=["Series"])
```

```python
# lambda series: f"https://en.wikipedia.org/wiki/Star_Trek:_{series.replace(' ','_')}"
df['Link'] = df['Series'].apply(lambda series: f"https://en.wikipedia.org/wiki/Star_Trek:_{series.replace(' ','_')}")
df
```
